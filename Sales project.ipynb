```python
#grouping by category(showing sum of sales in each category)
sales_by_category=df.groupby('Category')[['Sales']].sum().reset_index()
```

```python
#grouping category by sales
sales_by_category
```

```python
#PLOTTING A BAR CHAT OF CATEGORY AGAINST SALES
plt.figure(figsize=(5,6))
plt.bar(sales_by_category['Category'],sales_by_category['Sales'],color='green')
plt.title('SALES BY CATEGORY')
plt.xlabel('CATEGORY')
plt.ylabel('TOTAL')
plt.xticks(rotation = 45)
plt.show()
```

```python
#grouping category by profit
Profit_by_category=df.groupby('Category')[['Profit']].sum().reset_index()
Profit_by_category
```

```python
#PLOTTING A BAR CHAT OF CATEGORY AGAINST PROFIT
plt.figure(figsize=(5,6))
plt.bar(Profit_by_category['Category'],Profit_by_category['Profit'],color='green')
plt.title('PROFIT BY CATEGORY')
plt.xlabel('CATEGORY')
plt.ylabel('TOTAL')
plt.xticks(rotation = 45)
plt.show()
```

```python
#grouping region by sales
sales_by_Region=df.groupby('Region')[['Sales']].sum().reset_index()
sales_by_Region
```

```python
#plotting a barchat of region by sales
plt.figure(figsize=(2,3))
plt.bar(sales_by_Region['Region'],sales_by_Region['Sales'],color='green')
plt.title('SALES BY REGION')
plt.xlabel('REGION')
plt.ylabel('TOTAL')
plt.xticks(rotation = 45)
plt.show()
```

```python
#grouping region by profit
Profit_by_Region=df.groupby('Region')[['Profit']].sum().reset_index()
Profit_by_Region
```

```python
#plotting pie chart of region by profit
plt.figure(figsize=(5,7))
plt.pie(Profit_by_Region['Profit'], labels =Profit_by_Region['Region'] , startangle = 90)
plt.title('PROFIT BY REGION')
plt.show()
```

```python
#grouping region by category and sales
df_grouped = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
df_grouped
```

```python
# Create the bar plot with the sum of sales on the y-axis
ax = sns.barplot(data=df_grouped, x='Region', y='Sales', hue='Category')

# Add labels to the bars
#for bars in ax.containers:
 #   ax.bar_label(bars)

# Show the plot
plt.show()
```
